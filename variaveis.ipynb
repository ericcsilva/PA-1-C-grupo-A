```python
# Entrada
num1 = int (input("digite um numero: "))
num2 = int (input("digite um numero: "))

# Processamento
igual = num1 == num2 #igual
diferente = num1 != num2 #diferente
maior = num1 > num2 # maior
menor = num1 < num2 # menor
menorigual = num1 <= num2 # menor ou igual
maiorigual = num1 >= num2 # maior e igual

# Saida
print (igual)
print (diferente)
print (maior)
print (menor)
print (menorigual)
print (maiorigual)
```
